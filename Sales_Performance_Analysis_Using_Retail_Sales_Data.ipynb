# Check Dataset Structure

```python
# Check dataset information
df.info()

# Missing values
print(df.isnull().sum())

# Duplicate rows
print("Duplicate Rows:", df.duplicated().sum())
```

# Handle Missing Values and Duplicates

```python
# Remove duplicates
df = df.drop_duplicates()

# Fill missing values
for col in df.select_dtypes(include='object'):
    df[col].fillna(df[col].mode()[0], inplace=True)

for col in df.select_dtypes(include=['int64','float64']):
    df[col].fillna(df[col].median(), inplace=True)

print(df.isnull().sum())
```

# Rename Columns

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_')

print(df.columns)
```

# Convert Date Column and Feature Engineering

```python
df['order_date'] = pd.to_datetime(df['order_date'])

df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month_name()
df['quarter'] = df['order_date'].dt.quarter

df.head()
```

# Outlier and Skewness Check

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.boxplot(x=df['sales'])
plt.title("Outlier Detection - Sales")
plt.show()

print("Skewness:", df['sales'].skew())
```

# **EDA and Visualizations**

# Univariate Analysis (Sales Distribution)
# Cell

```python
plt.figure(figsize=(8,5))
sns.histplot(df['sales'], bins=30, kde=True)

plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

# Interpretation

Features Used: Sales

# Insights
Sales distribution is right-skewed.
Most orders are small or medium-sized.
High-value orders are limited.
Revenue is influenced by a few large transactions.
Business should identify factors behind high-value orders.

# Sales by Region

```python
plt.figure(figsize=(8,5))

region_sales = df.groupby('region')['sales'].sum().sort_values(ascending=False)

sns.barplot(x=region_sales.index,
            y=region_sales.values)

plt.title("Total Sales by Region")
plt.xticks(rotation=45)
plt.show()
```

# Interpretation

Features Used: Region, Sales

# Insights
Some regions contribute significantly more revenue.
Regional performance varies considerably.
Top-performing regions drive business growth.
Low-performing regions need strategic focus.
Regional marketing strategies can be optimized.

# Profit vs Sales


```python
plt.figure(figsize=(8,5))

sns.scatterplot(data=df,
                x='sales',
                y='profit')

plt.title("Sales vs Profit")
plt.show()
```

# Interpretation

Features Used: Sales, Profit

# Insights
Higher sales generally increase profit.
Some high-sales orders have low profit.
Discounts may reduce profitability.

Profit optimization is important.
